```python
import  pandas as pd
```

```python
pd.set_option('display.max_columns',100)
```

```python
df = pd.read_csv('02 DB Commodities/data.csv')
```

```python
df.head()
```

```python
df=df.drop(columns=['Date'])
df.head()
```

```python
x=df.drop(columns=['coffe'])
y=df['coffe']

```

```python
x.head()

```

```python
y.head()
```

```python
#Model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
#Predictions
```

```python
predictions = model.predict(x)
```

```python
predictions
```

```python
actuals = y.values
```

```python
actuals
```

```python
mae=0
for i in range(0, len(predictions)):
    prediction=predictions[i]
    actual = actuals[i]
    abs_error = abs(prediction-actual)
    mae = mae + abs_error

    #print(prediction, " - ",actual)
mae=mae/len(predictions)

```

```python
print('mae: ',mae)
print('mean: ',actuals.mean())
print('mae mean ratio: ',100*mae/actuals.mean(),'%')
```

```python
#Equation
```

```python
intercept = model.intercept_
```

```python
intercept
```

```python
coefficients = model.coef_
```

```python
coefficients
```

```python
print('y = ',intercept,' + ',end='')

for i in range(0,len(coefficients)):
    print(coefficients[i],'',x.columns[i],' + ', end ='')

print((' E'))
```

```python
# beta1 -> dollars/ number of bedrooms
# beta1*stdev(number of ) -> dollars

```

```python
features = pd.DataFrame(coefficients, x.columns, columns=['coefficient'])
```

```python
features.coefficient = features.coefficient.abs()
```

```python
features
```

```python
#Standar Deviations
```

```python
stdevs = []
for i in x.columns:
    stdev = df[i].std()
    print(i,"->",stdev)
    stdevs.append(stdev)
```

```python
stdevs
```

```python
import numpy as np
```

```python
features['std']=np.array(stdevs).reshape(-1,1)
features['importance'] = features['std']*features['coefficient']
features['importance_normalized'] = 100*features['importance']/features['importance'].max()
```

```python
features
features.sort_values(by='importance_normalized')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(features.index,features.importance_normalized)
```
